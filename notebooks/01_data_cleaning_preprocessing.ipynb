```python
# =========================================================
# 1. IMPORT REQUIRED LIBRARIES
# =========================================================

import pandas as pd
import numpy as np

from sklearn.impute import KNNImputer
```

```python
# =========================================================
# 2. LOAD SPSS DATASET
# =========================================================

# Dataset with categorical labels
df_cat = pd.read_spss(
    "KR_selected_child_vaccination.sav"
)

# Dataset with original numeric codes
df = pd.read_spss(
    "KR_selected_child_vaccination.sav",
    convert_categoricals=False
)
```

```python
# =========================================================
# 3. VARIABLE RENAMING
# =========================================================

col_names = {

    # Vaccination Variables
    "H1": "health_card",
    "H2": "bcg",
    "H3": "dpt1",
    "H4": "polio1",
    "H5": "dpt2",
    "H6": "polio2",
    "H7": "dpt3",
    "H8": "polio3",
    "H9": "measles1",
    "H51": "pentavalent1",
    "H52": "pentavalent2",
    "H53": "pentavalent3",

    # Demographic Variables
    "V024": "state",
    "V025": "residence_type",
    "V013": "age_group_5yr",
    "V190": "wealth_index",

    # Household Variables
    "V124": "has_motorcycle",
    "V125": "has_car_truck",

    # Education Variables
    "V106": "highest_educational_level",
    "V107": "highest_year_of_education",

    # Respondent Characteristics
    "V012": "respondent_age",
    "V157": "reads_newspaper",
    "V158": "listens_radio",
    "V159": "watches_tv",
    "V151": "household_sex_female",
    "V463Z": "no_tobacco_use",

    # Maternal & Delivery Variables
    "M14": "antenatal_visits",
    "M15": "place_of_delivery",

    # Child Variables
    "B4": "child_sex",
    "BORD": "birth_order",
    "B19": "child_age_months",
    "B5": "child_alive",

    # Sampling Variables
    "V005": "sample_weight",
    "V021": "psu",
    "V022": "strata",

    # Health Access Variables
    "V467D": "distance_health_facility_problem",

    # ICDS Variables
    "S558": "received_icds_benefits",
    "S361": "met_community_health_worker",

    # Media Exposure Variables
    "V384A": "fp_radio_exposure",
    "V384B": "fp_tv_exposure",
    "V384C": "fp_newspaper_exposure",

    # Health Variables
    "M17": "c_section_delivery",
    "HW57": "child_anemia_level",
    "V457": "woman_anemia_level",
    "M2A": "prenatal_doctor_visit",
    "M2B": "prenatal_nurse_visit"
}
```

```python
# =========================================================
# 4. APPLY VARIABLE RENAMING
# =========================================================

df = df.rename(columns=col_names)

df_cat = df_cat.rename(columns=col_names)
```

```python
# =========================================================
# 5. SELECT REQUIRED VARIABLES
# =========================================================

selected_columns = [

    # Vaccination Variables
    'bcg',
    'dpt1',
    'polio1',
    'dpt2',
    'polio2',
    'dpt3',
    'polio3',
    'measles1',
    'pentavalent1',
    'pentavalent2',
    'pentavalent3',

    # Demographic Variables
    'state',
    'residence_type',
    'age_group_5yr',
    'wealth_index',

    # Household Variables
    'has_motorcycle',
    'has_car_truck',

    # Education Variables
    'highest_educational_level',
    'highest_year_of_education',

    # Maternal Variables
    'respondent_age',
    'antenatal_visits',
    'place_of_delivery',

    # Child Variables
    'child_sex',
    'birth_order',
    'child_age_months',
    'child_alive',

    # Media Exposure
    'reads_newspaper',
    'listens_radio',
    'watches_tv',

    # Household Characteristics
    'household_sex_female',
    'no_tobacco_use',

    # ICDS Variables
    'received_icds_benefits',

    # Survey Variables
    'sample_weight',
    'psu',
    'strata',

    # Healthcare Access
    'distance_health_facility_problem',
    'met_community_health_worker',

    # Family Planning Exposure
    'fp_radio_exposure',
    'fp_tv_exposure',
    'fp_newspaper_exposure',

    # Health Variables
    'c_section_delivery',
    'child_anemia_level',
    'woman_anemia_level',

    # Prenatal Care
    'prenatal_doctor_visit',
    'prenatal_nurse_visit'
]

df = df[selected_columns]
```

```python
# =========================================================
# 6. INITIAL DATA INSPECTION
# =========================================================

print("Dataset Shape:", df.shape)

df.head()
```

```python
# =========================================================
# 7. DATASET INFORMATION
# =========================================================

df.info()
```

```python
# =========================================================
# 8. MISSING VALUE ANALYSIS
# =========================================================

df.isnull().sum()
```

```python
# =========================================================
# 9. OUTLIER DETECTION USING IQR METHOD
# =========================================================

num_cols = [

    'respondent_age',
    'antenatal_visits'
]

outlier_indices = set()

for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers_col = df[
        (df[col] < lower) |
        (df[col] > upper)
    ].index

    outlier_indices.update(outliers_col)

outlier_indices = list(outlier_indices)

print("Total Outlier Rows:", len(outlier_indices))
```

```python
# =========================================================
# 10. REMOVE OUTLIERS
# =========================================================

df = df.drop(index=outlier_indices)

df = df.reset_index(drop=True)

print("Shape After Outlier Removal:", df.shape)
```

```python
# =========================================================
# 11. VARIABLE RECODING
# =========================================================

# Male = 0
# Female = 1

df['household_sex_female'] = (

    df['household_sex_female']
    .map({1: 0, 2: 1})
)

# Distance to health facility

df['distance_health_facility_problem'] = (

    df['distance_health_facility_problem']
    .map({0: 0, 1: 2, 2: 1})
)
```

```python
# =========================================================
# 12. HANDLE INVALID / SPECIAL CODES
# =========================================================

# Replace 98 with missing value
df["antenatal_visits"] = (

    df["antenatal_visits"]
    .replace(98, np.nan)
)

# Replace 7 with missing value

cols = [

    'has_motorcycle',
    'has_car_truck'
]

df[cols] = df[cols].replace(7, np.nan)
```

```python
# =========================================================
# 13. STATE-WISE MEDIAN IMPUTATION FUNCTION
# =========================================================

def statewise_median_imputation(dataframe, column):

    dataframe[column] = (

        dataframe
        .groupby("state")[column]
        .transform(
            lambda x: x.fillna(x.median())
        )
    )

    return dataframe
```

```python
# =========================================================
# 14. APPLY STATE-WISE MEDIAN IMPUTATION
# =========================================================

impute_cols = [

    "antenatal_visits",
    "highest_year_of_education",
    "child_anemia_level",
    "woman_anemia_level",
    "prenatal_doctor_visit",
    "prenatal_nurse_visit",
    "has_motorcycle",
    "has_car_truck"
]

for col in impute_cols:

    df = statewise_median_imputation(df, col)
```

```python
# =========================================================
# 15. VERIFY MISSING VALUES
# =========================================================

df.isnull().sum()
```

```python
# =========================================================
# 16. DEFINE VACCINATION VARIABLES
# =========================================================

vaccine_vars = [

    "bcg",
    "dpt1",
    "polio1",
    "dpt2",
    "polio2",
    "dpt3",
    "polio3",
    "measles1",
    "pentavalent1",
    "pentavalent2",
    "pentavalent3"
]
```

```python
# =========================================================
# 17. IDENTIFY "DON'T KNOW" RESPONSES
# =========================================================

df["dont_know"] = (

    df[vaccine_vars]
    .isin([8])
    .any(axis=1)

).astype(int)

df["dont_know"].value_counts()
```

```python
# =========================================================
# 18. REMOVE "DON'T KNOW" OBSERVATIONS
# =========================================================

df = df[df["dont_know"] == 0]

print("Shape After Removing Don't Know:", df.shape)
```

```python
# =========================================================
# 19. CREATE FULLY VACCINATED TARGET VARIABLE
# =========================================================

# BCG
bcg = df["bcg"].isin([1, 2, 3])

# DPT Completion
dpt_complete = (

    df["dpt1"].isin([1, 2, 3]) &
    df["dpt2"].isin([1, 2, 3]) &
    df["dpt3"].isin([1, 2, 3])
)

# Pentavalent Completion
penta_complete = (

    df["pentavalent1"].isin([1, 2, 3]) &
    df["pentavalent2"].isin([1, 2, 3]) &
    df["pentavalent3"].isin([1, 2, 3])
)

# DPT or Pentavalent
dpt_or_penta = dpt_complete | penta_complete

# Polio Completion
polio_complete = (

    df["polio1"].isin([1, 2, 3]) &
    df["polio2"].isin([1, 2, 3]) &
    df["polio3"].isin([1, 2, 3])
)

# Measles
measles = df["measles1"].isin([1, 2, 3, 4])

# Final Vaccination Variable
df["vaccinated"] = (

    bcg &
    dpt_or_penta &
    polio_complete &
    measles

).astype(int)
```

```python
# =========================================================
# 20. KNN IMPUTATION
# =========================================================

cols_to_impute = [

    'has_motorcycle',
    'has_car_truck'
]

features = cols_to_impute + ['vaccinated']

knn_df = df[features].copy()

imputer = KNNImputer(

    n_neighbors=5,
    weights='distance'
)

imputed = imputer.fit_transform(knn_df)

imputed_df = pd.DataFrame(

    imputed,
    columns=features,
    index=df.index
)

for col in cols_to_impute:

    df[col] = (

        imputed_df[col]
        .round()
        .astype(int)
    )
```

```python
# =========================================================
# 21. ANTENATAL VISITS INDICATOR
# =========================================================

# 1 = 4 or more ANC visits
# 0 = Less than 4 visits

df["antenatal_visits_yes"] = np.where(

    (
        (df["antenatal_visits"] >= 4) &
        (df["antenatal_visits"] < 98)
    ),

    1,
    0
)
```

```python
# =========================================================
# 22. INSTITUTIONAL DELIVERY
# =========================================================

# Codes 20–33 correspond to institutional delivery

df["institutional_delivery"] = (

    df["place_of_delivery"]
    .isin(list(range(20, 34)))

).astype(int)
```

```python
# =========================================================
# 23. HIGH BIRTH ORDER
# =========================================================

# 1 = Birth order > 5

df["birth_order_gt_5"] = np.where(

    df["birth_order"] > 5,
    1,
    0
)
```

```python
# =========================================================
# 24. URBAN RESIDENCE
# =========================================================

# 1 = Urban
# 0 = Rural

df["residence_urban"] = (

    df["residence_type"] == 1

).astype(int)
```

```python
# =========================================================
# 25. MATERNAL AGE > 35 YEARS
# =========================================================

df["mothers_age_gt_35"] = np.where(

    df["respondent_age"] > 35,
    1,
    0
)
```

```python
# =========================================================
# 26. MEDIA EXPOSURE
# =========================================================

newspaper = df["reads_newspaper"].isin([1, 2, 3])

radio = df["listens_radio"].isin([1, 2, 3])

tv = df["watches_tv"].isin([1, 2, 3])

df["media_exposure"] = (

    newspaper |
    radio |
    tv

).astype(int)
```

```python
# =========================================================
# 27. HIGHER EDUCATION STATUS
# =========================================================

# 1 = Secondary/Higher Education

df["education_status_higher"] = (

    df["highest_educational_level"]
    .isin([2, 3])

).astype(int)
```

```python
# =========================================================
# 28. RICH WEALTH STATUS
# =========================================================

# 1 = Richer/Richest

df["wealth_status_rich"] = (

    df["wealth_index"]
    .isin([4, 5])

).astype(int)
```

```python
# =========================================================
# 29. ICDS BENEFITS
# =========================================================

df["recieves_icds_benefits"] = (

    df["received_icds_benefits"]
)
```

```python
# =========================================================
# 30. HEALTH FACILITY DISTANCE ISSUE
# =========================================================

df["facility_distance_issue"] = (

    df["distance_health_facility_problem"]
    .isin([1, 2])

).astype(int)
```

```python
# =========================================================
# 31. AUTOMOBILE OWNERSHIP
# =========================================================

motor = df["has_motorcycle"].isin([1])

car = df["has_car_truck"].isin([1])

df["has_automobile"] = (

    motor |
    car

).astype(int)
```

```python
# =========================================================
# 32. COMMUNITY HEALTH WORKER VISIT
# =========================================================

df["health_worker_visit"] = (

    df["met_community_health_worker"]
)
```

```python
# =========================================================
# 33. FAMILY PLANNING EXPOSURE SCORE
# =========================================================

df["fp"] = df[

    [
        "fp_radio_exposure",
        "fp_tv_exposure",
        "fp_newspaper_exposure"
    ]

].sum(axis=1)
```

```python
# =========================================================
# 34. REMOVE DUPLICATE OBSERVATIONS
# =========================================================

# Check duplicate rows
print("Duplicate Rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Reset index
df = df.reset_index(drop=True)

# Verify dataset shape
print("Dataset Shape After Removing Duplicates:")

print(df.shape)
```

```python
# =========================================================
# 35. FINAL DATASET SUMMARY
# =========================================================

print("Final Dataset Shape:", df.shape)

print("\nVaccination Status Distribution:\n")

print(df["vaccinated"].value_counts())

print("\nRemaining Missing Values:\n")

print(df.isnull().sum().sum())
```

```python
# =========================================================
# 36. SAVE CLEAN DATASET
# =========================================================

df.to_csv(

    "clean_child_vaccination_data.csv",
    index=False
)

print("Clean dataset saved successfully.")
```
